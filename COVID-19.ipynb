**Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Data Cleaning**

```python
df.columns = ['Region','Date','Frequency','Estimated_Unemployment_Rate',
              'Estimated_Employed','Estimated_Labour_Participation_Rate',
              'Area','Latitude','Longitude']
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

**Check Missing Values**

```python
df.isnull().sum()
```

```python
# Drop missing values if any
df.dropna()
```

**Feature Engineering**

```python
# Extract Year and Month

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

df.head()
```

**Exploratory Data Analysis (EDA)**

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Estimated_Unemployment_Rate'], kde=True)
plt.title("Distribution of Unemployment Rate")
plt.show()
```

**Unemployment Rate by Region**

```python
plt.figure(figsize=(10,8))
sns.barplot(x='Estimated_Unemployment_Rate', y='Region', data=df)
plt.title("Unemployment Rate by Region")
plt.show()
```

**Area Wise Analysis (Urban vs Rural)**

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Area', y='Estimated_Unemployment_Rate', data=df)
plt.title("Urban vs Rural Unemployment Rate")
plt.show()
```

**Covid-19 Unemployment Rate**

```python
yearly_data = df.groupby('Year')['Estimated_Unemployment_Rate'].mean().reset_index()
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(data=yearly_data, x='Year', y='Estimated_Unemployment_Rate', marker='o')

plt.title("Year Wise Average Unemployment Rate")
plt.xlabel("Year")
plt.ylabel("Average Unemployment Rate (%)")

plt.xticks(yearly_data['Year'])
plt.grid(True)

plt.show()
```

**Monthly Unemployment Trend**

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Month', y='Estimated_Unemployment_Rate', data=df)
plt.title("Monthly Unemployment Trend")
plt.show()
```

**Heatmap of Unemployment Rate**

```python
plt.figure(figsize=(14,8))
pivot = df.pivot_table(
    values='Estimated_Unemployment_Rate',
    index='Region',
    columns='Month',
    aggfunc='mean'
)
sns.heatmap(
    pivot,
    annot=True,
    fmt=".1f",
    cmap='YlOrRd',
    linewidths=0.5
)
plt.title("Monthly Unemployment Rate by Region", fontsize=16)
plt.xlabel("Month")
plt.ylabel("Region")

plt.show()
```

**Correlation Analysis**

```python
plt.figure(figsize=(10,4))
sns.heatmap(df[['Estimated_Unemployment_Rate','Estimated_Employed',
                'Estimated_Labour_Participation_Rate']].corr(),
            annot=True)

plt.title("Correlation Analysis")
plt.show()
```
